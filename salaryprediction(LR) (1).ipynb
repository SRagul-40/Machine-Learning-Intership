```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
saldf=pd.read_csv("/content/drive/MyDrive/Salary Data.csv")
```

```python
saldf.head()
```

```python
saldf.info()
```

```python
saldf.isnull().sum()
```

```python
saldf=saldf.dropna()
```

```python
inp=saldf[['Years of Experience']]
out=saldf['Salary']
```

```python
LR=LinearRegression()
LR.fit(inp,out)
```

```python
LR.predict([[10]])
```

```python
predicted_values = LR.predict(inp)

plt.figure(figsize=(10,6))
plt.scatter(out, predicted_values, alpha=0.5,label='Actual vs Predicted')

# Add a line representing perfect prediction (y=x)
min_val = min(out.min(), predicted_values.min())
max_val = max(out.max(), predicted_values.max())
plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--', label='Perfect Prediction')
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.title("Actual vs Predicted Salary")
plt.legend()
plt.grid(True)
plt.show()
```
